```python
# Import Libraries
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import string
import csv
import re
```

```python
# Open CSV File & Cleaning
listwit = []
with open('tweetutf2.csv',encoding='utf-8') as data:
    reader = csv.reader(data)
    for row in reader:
        row[0] = re.sub(r'\s\s+',' ', row[0])
        row[0] = re.sub(r'http\S+', '', row[0]) #hapus link http
        row[0] = re.sub(r'@\S+', '', row[0]) #hapus mention
        row[0] = re.sub(r'Membalas', '', row[0]) #hapus membalas
        row[0] = re.sub(r'RT', '', row[0]) #hapus RT
        row[0] = re.sub(r"\.", '', row[0]) #hapus titik
        row[0] = re.sub(r",", '', row[0]) #hapus koma
        row[0] = re.sub(r":", '', row[0]) #hapus titik dua
        row[0] = re.sub(r"\?", '', row[0]) #hapus tanda tanya
        row[0] = re.sub(r"!", '', row[0]) #hapus tanda seru
        row[0] = re.sub(r"#", '', row[0]) #hapus hashtag
        row[0] = re.sub(r"\(", '', row[0]) #hapus (
        row[0] = re.sub(r"\)", '', row[0]) #hapus )
        row[0] = re.sub(r"\[", '', row[0]) #hapus [
        row[0] = re.sub(r"\]", '', row[0]) #hapus ]
        row[0] = re.sub(r"[^A-Za-z0-9().,<_>!?\'`]", ' ', row[0]) #hapus entah apa ini :v
        row[0] = re.sub(r"\s{2,}", ' ', row[0]) #hapus entah apa ini :v (2)
        listwit.append(row[0].split(';'))
listwit
```
